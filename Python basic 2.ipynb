```python
#argument passing
def evenodd (x):
    if x% 2==0:
        print("even")
    else:
        print("odd")
evenodd(10)
evenodd(15)
```

```python
#default argument
def myfun(x,y=20):
    print("x:",x)
    print("y:",y)
myfun(30)
```

```python
#key argument
def student (fname, lname):
    print(fname, lname)

student(fname="Practice", lname="Daily")
student(fname="Daily", lname="Practice")
```

```python
#positional argument
def person(name,age):
    print("Hi I am",name)
    print("My age is",name)

#case1
person("Ram", 25)

#case2
person("Ram", 25)
```

```python
#function within function
def f1():
    s="I love you"
    def f2():
        print(s)
    f2()
f1()    
```

```python
#pass by refernce
def myfun(x):
    x[0]=20
lst=[10,20,30,40,50]
print(lst)
myfun(lst)
print(lst)
```

```python
#pass by value
def myFun(x):
    x=20
a=20
print(a)
myFun(a)
print(a)
#a is not modified
```

```python
#writing file
with open("geek.txt", "w") as file:
    file.write("Hello Python!\n")
    file.write("File Handling is easy with Pyrthon.")
print("File written successfully")
```

```python
# using with statement
with open("geek.txt", "r") as file:
    content=file.read()
    print(content)
```

```python
# handling exceptions
try:
    file=open("geek.txt", "r")
    content = file.read()
    print(content)
finally:
    file.close()
```

```python
# Exception Handling
# basic example
n=10
try:
    res=n/0
except ZeroDivisionError:
    print("Can't be divided by Zero!")
```

```python
try:
    n=0
    res=100/n

except ZeroDivisionError:
    print("You can't divided by Zero!") 

except ValueError:
    print("Enter a valid number!")

else:
    print("Result is", res)

finally:
    print("Execution complete.")
    
```

```python
#catching specific exceptionas
try:
    x=int("str")
    inv=1/x

except ValueError:
    print("Not Valid!")

except ZeroDivisionError:
    print("Zero has no inverse!")
```

```python
# catching multiple exceptions
a=["10", "twenty", 30]

try:
    total=int(a[0]) + int(a[1])

except (ValueError, TypeError) as e:
    print("Error", e)

except IndexError:
    print("Index out of range")
```

```python
# catch all handlers
try:
    res="100"/20

except ArithmeticError:
    print("Arithmetic problem.")

except:
    print("Something went wrong!")
```

```python
# raise an exception
def set(age):
    if age<0:
        raise ValueError("Age cannot be negative")
    print(f"Age set to {age}")

try:
    set(-5)
except ValueError as e:
    print(e)
```
